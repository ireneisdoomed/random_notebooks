```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
IPythonConsole.ipython_useSVG=True  #< set this to False if you want PNGs instead of SVGs
```

### Paclitaxel

```python
# SMILES in ChEMBL (Open Targets) that doesn't match the one in ChemSpider and PubChem
chembl = 'CC(=O)O[C@H]1C(=O)[C@]2(C)[C@@H](O)C[C@H]3OC[C@@]3(OC(C)=O)[C@H]2[C@H](OC(=O)c2ccccc2)[C@]2(O)C[C@H](OC(=O)[C@H](O)[C@@H](NC(=O)c3ccccc3)c3ccccc3)C(C)=C1C2(C)C'

Chem.MolFromSmiles(chembl)
```

```python
# SMILES in ChemSpider - doesn't match with the structure in their page. Chirality is different.
chemspider = 'CC1=C2[C@H](C(=O)[C@@]3([C@H](C[C@@H]4[C@]([C@H]3[C@@H]([C@@](C2(C)C)(C[C@@H]1OC(=O)[C@@H]([C@H](c5ccccc5)NC(=O)c6ccccc6)O)O)OC(=O)c7ccccc7)(CO4)OC(=O)C)O)C)OC(=O)C'

Chem.MolFromSmiles(chemspider)
```

```python
# SMILES in PubChem - Canonical representation seems to be the same one as the one from chemspider above
pubchem_canonical = 'CC1=C2[C@H](C(=O)[C@@]3([C@H](C[C@@H]4[C@]([C@H]3[C@@H]([C@@](C2(C)C)(C[C@@H]1OC(=O)[C@@H]([C@H](C5=CC=CC=C5)NC(=O)C6=CC=CC=C6)O)O)OC(=O)C7=CC=CC=C7)(CO4)OC(=O)C)O)C)OC(=O)C'
pubchem_isomeric = 'CC1=C2C(C(=O)C3(C(CC4C(C3C(C(C2(C)C)(CC1OC(=O)C(C(C5=CC=CC=C5)NC(=O)C6=CC=CC=C6)O)O)OC(=O)C7=CC=CC=C7)(CO4)OC(=O)C)O)C)OC(=O)C'

Chem.MolFromSmiles(pubchem_canonical)
```

```python
# The isomeric SMILES is supposed to capture the stereochemistry, but still shows an isomer of the structure of ChemSpider

Chem.MolFromSmiles(pubchem_isomeric)
```

```python
# DrugBank's representation coincides with that of ChemSpider but when plotting their smiles, they show the enantiomer

drugbank = '[H][C@]12[C@H](OC(=O)C3=CC=CC=C3)[C@]3(O)C[C@H](OC(=O)[C@H](O)[C@@H](NC(=O)C4=CC=CC=C4)C4=CC=CC=C4)C(C)=C([C@@H](OC(C)=O)C(=O)[C@]1(C)[C@@H](O)C[C@H]1OC[C@@]21OC(C)=O)C3(C)C'

Chem.MolFromSmiles(drugbank)
```
